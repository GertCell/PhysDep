```python
import numpy as np
from matplotlib.pyplot import *

%matplotlib notebook

X = []
Y = []
G = []
T = []

n = 25

for i in range(0, n):
    x = - 5 + i * 10 / n
    y = 1 / (1 + x ** 2)
    X.append(x)
    Y.append(y)

A = np.copy(Y)

for j in range(1, n):
    
    for i in range(n-1, j-1, -1):
        A[i] = float(A[i] - A[i - 1]) / float(X[i] - X[i - j])

def p(g):
    
    P = A[0]
    t = (g - X[0])
    
    for j in range(1, n):
        P = P + t * A[j]
        t = t * (g - X[j])
    return P

subplot(111)
axis([-5, 5, 0, 2])
ln1, = plot(G, T)

for i in range(0, n):
    scatter(X[i], Y[i])
    gcf().canvas.draw()

for x_2 in np.arange(-5, 5, 0.1):
    G.append(x_2)
    T.append(p(x_2))
    
    ln1.set_data(G, T)
    gcf().canvas.draw()
```

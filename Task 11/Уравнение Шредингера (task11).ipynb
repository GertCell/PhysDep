```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal

%matplotlib inline
```

```python

L = 10
n = 1000
a = -L
b = L
h = (b-a)/n

U = lambda x: -x**2/2

X = np.linspace(a,b,n+1)
V = np.linspace(1,2,n+1)
Vp = np.linspace(1,2,n+1)

H = np.zeros((n+1,n+1))


def tridiag():
    H[[0],[0]] =  1.0/h**2 + U(X[0])
    H[[0],[1]] = -(0.5/h**2)
    H[[n],[n-1]] = -(0.5/h**2)
    H[[n],[n]] = 1/h**2 + U(X[n])
    for i in range(1,n):
            H[[i],[i-1]] = -(0.5/h**2)
            H[[i],[i]] = 1.0/h**2 + U(X[i])
            H[[i],[i+1]] = -(0.5/h**2)
    return H

tridiag()

for i in range(1, 5):
    V = np.linalg.inv(H) @ Vp
    V = V/np.linalg.norm(V)
    Vp = V
    vec = H@V
    E = (V@vec) / (V@V)
    plt.plot(X,Vp)
    plt.gcf().canvas.draw()


vec = H@V
E = (V@vec) / (V@V)

print(E)
plt.plot(X,Vp)
plt.grid()
```

```python
from matplotlib.pyplot import *
import numpy as np
%matplotlib notebook

a = 10
b = 2
c = 2
d = 10

fx = lambda t, x, y: a * x - b * x * y

fy = lambda t, x, y: c * x * y - d * y

X = []
Y = []
T = []

T_h = 3
N = 100
h = T_h / N

t0 = 0
x0 = 1
y0 = 1

subplot(111)
axis([0,3, 0, 1])
ln1, = plot(T, X)
ln2, = plot(T, Y)

def rk2(f1, f2, x, y, t):
    
    for i in range(0, N):
        
        t_t_x = x + h * f1(t, x, y)
        x = x + h / 2 * f1(t, x, y) + h / 2 * f1(t + h, t_t_x, y + h)
        
        t_t_y = y + h * f2(t, x, y)
        y = y + h / 2 * f2(t, x, y) + h / 2 * f2(t + h, x + h,  t_t_y)
        
        t = t + h
        
        T.append(t)
        X.append(x)
        Y.append(y)
    ln1.set_data(T, X)
    ln2.set_data(T, Y)
    gcf().canvas.draw()
    return 0

rk2(fx, fy, x0, y0, t0)

```

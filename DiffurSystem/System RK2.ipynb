```python
from matplotlib.pyplot import *
import numpy as np

%matplotlib inline


a = 10
b = 2
c = 2
d = 10

fx = lambda t, x, y: a * x - b * x * y

fy = lambda t, x, y: c * x * y - d * y

X = []
Y = []
T = []

T_h = 2
N = 1000
h = T_h / N

t0 = 0
x0 = 6
y0 = 6
figure(figsize = (4,4), dpi = 150)
matplotlib.rc('xtick', labelsize=5)
matplotlib.rc('ytick', labelsize=5)
grid()

subplot(211)

axis([0, 2, 2, 8])
ln1, = plot(T, X)
ln2, = plot(T, Y)

tight_layout()

figure(figsize = (4,4), dpi = 150)
grid()

subplot(212)
axis([3.5, 6.5, 2, 8])
ln3, = plot(X, Y)

tight_layout()

def rk2(f1, f2, x, y, t):
    
    for i in range(0, N):
        
        t_t_x = x + h * f1(t, x, y)
        t_t_y = y + h * f2(t, x, y)
        
        x = x + h / 2 * f1(t, x, y) + h / 2 * f1(t + h, t_t_x, t_t_y)     
        y = y + h / 2 * f2(t, x, y) + h / 2 * f2(t + h, t_t_x,  t_t_y)
        
        X.append(x)
        Y.append(y)
        
    T = np.arange(0, T_h, h)
    
    ln1.set_data(T, X)
    ln2.set_data(T, Y)
    ln3.set_data(X, Y)
    
    gcf().canvas.draw()
    return 0

rk2(fx, fy, x0, y0, t0)

# особые точки задачи, период вблизи точки (5,5)
```

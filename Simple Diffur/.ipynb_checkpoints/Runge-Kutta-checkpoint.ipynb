```python
from scipy import *
from matplotlib.pyplot import *
import numpy as np
%matplotlib notebook

f_1 = lambda t, x: -x

T_h = 3
N = 1000
h = T_h / N
t0 = 0
x0 = 1

T_1 = []
X_1 = []
T_1_err = []

T_2 = []
X_2 = []
T_2_err = []

T_4 = []
X_4 = []
T_4_err = []

X = np.arange(0, T_h, h)

subplot(211)
axis([0, 3, 0, 1])

xlabel('T')
ylabel('X')
title('Ð¡omparison of Runge-Kutta methods')

ln, = plot(X, np.exp(-X), label = 'Exp(-t)')
ln1, = plot(X_1, T_1, label = 'Runge-Kutta 1')
ln2, = plot(X_2, T_2, label = 'Runge-Kutta 2')
ln4, = plot(X_4, T_4, label = 'Runge-Kutta 4')

legend()
grid()

subplot(212)
axis([0,2, 0, 0.05])
ln_1, = plot([], T_1_err, label = 'err Runge-Kutta 1')
ln_2, = plot([], T_2_err, label = 'err Runge-Kutta 2')
ln_4, = plot([], T_4_err, label = 'err Runge-Kutta 4')
legend()
grid()

def coefs(f, x, t):
    k1 = f(x, t)
    k2 = f(x + h / 2, t + h / 2 * k1)
    k3 = f(x + h / 2, t + h / 2 * k2)
    k4 = f(x + h, t + h * k3)
    return (k1 + 2 * k2 + 2 * k3 + k4) * h / 6

def rk1(f, x, t):
    for i in range(0, N):
        t = t + h * f(x, t)
        
        x = x + h
        T_1.append(t)
        X_1.append(x)
        T_1_err.append(np.exp(-x) - t)
        ln1.set_data(X_1,T_1)
        ln_1.set_data(X_1, T_1_err)
    gcf().canvas.draw()
    return 0

def rk2(f, x, t):
    alpha = 3/4
    for i in range(0, N):
        
        t = t + h * ((1-alpha) * f(x, t) + alpha * f(x + h/2/alpha, t + h * f(x, t)))
        
        
        
        x = x + h
        T_2_err.append(np.exp(-x) - t)
        T_2.append(t)
        X_2.append(x)
        
        ln_2.set_data(X_2, T_2_err)
        
        ln2.set_data(X_2, T_2)
    gcf().canvas.draw()
    return 0
        
def rk4(f, x, t):

    for i in range(0, N):
        t = t + coefs(f, x, t)
        
        x = x + h
        T_4_err.append(np.exp(-x) - t)
        T_4.append(t)
        X_4.append(x)
        
        ln_4.set_data(X_4, T_4_err)
        ln4.set_data(X_4,T_4)
    gcf().canvas.draw()
    return 0

rk1(f_1, t0, x0)
rk2(f_1, t0, x0)
rk4(f_1, t0, x0)



```

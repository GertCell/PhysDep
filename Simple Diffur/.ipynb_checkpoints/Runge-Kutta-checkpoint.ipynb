```python
from scipy import *
from matplotlib.pyplot import *
import numpy as np
%matplotlib notebook

f_1 = lambda t, x: -x

T_h = 3
N = 100
h = T_h / N
t0 = 0
x0 = 1

T_1 = []
X_1 = []

T_2 = []
X_2 = []

T_4 = []
X_4 = []

X = np.arange(0, T_h, h)

subplot(111)
axis([0, 3, 0, 1])

xlabel('T')
ylabel('X')
title('Ð¡omparison of Runge-Kutta methods')

ln, = plot(X, np.exp(-X), label = 'Exp(-t)')
ln1, = plot(X_1, T_1, label = 'Runge-Kutta 1')
ln2, = plot(X_2, T_2, label = 'Runge-Kutta 2')
ln4, = plot(X_4, T_4, label = 'Runge-Kutta 4')

legend()
grid()


def coefs(f, x, t):
    k1 = f(x, t)
    k2 = f(x + h / 2, t + h / 2 * k1)
    k3 = f(x + h / 2, t + h / 2 * k2)
    k4 = f(x + h, t + h * k3)
    return (k1 + 2 * k2 + 2 * k3 + k4) * h / 6

def rk1(f, x, t):
    for i in range(0, N):
        t = t + h * f(x, t)
        x = x + h
        T_1.append(t)
        X_1.append(x)
        ln1.set_data(X_1,T_1)
        gcf().canvas.draw()
    return 0

def rk2(f, x, t):
    for i in range(0, N):
        t_t = t + h * f(x, t)
        t = t + h / 2 * f(x, t) + h / 2 * f(x + h, t_t)
        x = x + h
        T_2.append(t)
        X_2.append(x)
        ln2.set_data(X_2,T_2)
        gcf().canvas.draw()
    return 0
        
def rk4(f, x, t):

    for i in range(0, N):
        t = t + coefs(f, x, t)
        x = x + h
        T_4.append(t)
        X_4.append(x)
        ln4.set_data(X_4,T_4)
        gcf().canvas.draw()
    return 0

rk1(f_1, t0, x0)
rk2(f_1, t0, x0)
rk4(f_1, t0, x0)



```

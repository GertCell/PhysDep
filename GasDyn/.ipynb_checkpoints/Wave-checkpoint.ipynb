```python
from matplotlib.pyplot import *
import numpy as np
%matplotlib notebook

N = 1000
Nx = 2000
H = 1
T = 1

a = 0.5
h =  H / Nx
t = T / N

d = a * t / h

X = np.arange(0, H, h)
U = [3] * len(X)
U[0] = 5
U1 = np.copy(U)
UU = []

subplot(111)
#axis([0, 1, 3, 5.01])
ln, = plot(X, U)
ln1, = plot(X, U1)

for time in np.arange(0, T, t):
    UU = np.copy(U)

    for i in range(1, Nx):
        U[i] = UU[i] + d*(UU[i - 1] - UU[i])

    ln.set_data(X, U)
    gcf().canvas.draw()

    y = 1
    U[y] = 5
    y += 1


for time in np.arange(0, T, t):
    
    UU = np.copy(U1)

    for i in range(1, Nx):
        U1[i] = UU[i] + d*(UU[i - 1] - UU[i])

    ln.set_data(X, U1)
    gcf().canvas.draw()

    y = 1
    U1[y] = 5
    y += 1

```

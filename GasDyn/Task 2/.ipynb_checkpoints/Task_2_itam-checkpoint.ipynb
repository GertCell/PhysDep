```python
def root(f, a, b):

    n = 100

    if f(a) * f(b) < 0:
        for i in range(1, n):

            if f((a + b) / 2) * f(b) < 0:
                a = (a + b) / 2
            else:
                b = (a + b) / 2
            # print('current number of iterations =', n)

    else:
        print('change range or function has not any roots')
    # print('root =', (a + b) / 2)
    return (a + b) / 2
```

```python
def exact(N):

    x = np.linspace(0, 1, N+1)

    a_0, x_0 = 0.5, 0.5
    aMref = 1

    po_0 = 1
    p_0 = 1
    u_0 = 1

    def a(x):
        return 0.5 + (1-0.5)*(1-x/0.5)**2

    def mach(m):
        return -a(x[i])/a_0+aMref/m*((1+(gamma-1)/2*m**2)/(1+1/2*(gamma-1)*aMref**2))**((gamma+1)/(2*(gamma-1)))

    def velocity(m):
        return u_0*m/aMref*((1+(gamma-1)/2*aMref**2)/(1+(gamma-1)/2*m**2))**(1/2)

    def density(m):
        return po_0*((1+(gamma-1)/2*aMref**2)/(1+(gamma-1)/2*m**2))**(1/(gamma-1))

    def pressure(m):
        return p_0*(density(m)/po_0)**gamma

    U = np.linspace(0, L, N+1)
    Po = np.linspace(0, L, N+1)
    P = np.linspace(0, L, N+1)

    M = np.linspace(0, L, N+1)

    for i in range(0, int(N/2)):
        M[i] = root(mach, 0.001, 1)
    M[int(N/2)] = 1
    for i in range(int((N+2)/2), N+1):
        M[i] = root(mach, 1, 50)
    return density(M)/density(M[0]), pressure(M)/pressure(M[0])*8, velocity(M)/velocity(M[0])*1.0237498

# plt.plot(x, density(M)/density(M[0]))
# plt.plot(x, po)
# plt.plot(x, velocity(M)/velocity(M[0])*1.0237498)
# plt.plot(x, u)
# plt.plot(x, pressure(M)/pressure(M[0])*8)
# plt.plot(x, p)
```

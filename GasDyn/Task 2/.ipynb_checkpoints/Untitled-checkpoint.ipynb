```python
#ÑÑ…ÐµÐ¼Ð° ÐœÐ°ÐºÐºÐ¾Ñ€Ð¼Ð°ÐºÐ°
import math as mt
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
def Makk(n, parameter):
    eps = 10e-4
    gamma = 1.4
    c_0 = 3
    L = 1.0
    a = 0
    b = L
    h = (b-a)/n
    kappa = 0.5
    #tau = 0.00000001
    #0.000000001
    #0.00000001


    #Ð¸Ð½Ð¸Ñ†Ð¸Ð°Ð»Ð¸Ð·Ð°Ñ†Ð¸Ñ#
    x = np.linspace(0, L, n+1)
    A = np.linspace(0, L, n+1)
    dA = np.linspace(0, L, n+1)
    e = np.linspace(0, L, n+1)

    po = np.linspace(0, L, n+1)
    poprev = np.linspace(0, L, n+1)
    p = np.linspace(0, L, n+1)
    M = []
    u = np.linspace(0, L, n+1)

    Q = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    F = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    S = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    Qnext = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    Qs = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    Fs = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    Ss = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])

    As = 0.5
    xs = 0.5
    for i in range(0,n+1):
        A[i] = As + (1-As)*(1-x[i]/xs)**2 
    for i in range(1,n):
        dA[i] = 0.5 * (A[i+1] - A[i-1])/h  #4*x[i] - 2   

    #Ð½Ð°Ñ‡ Ð¸ Ð³Ñ€Ð°Ð½ ÑƒÑÐ»Ð¾Ð²Ð¸Ñ#

    #=====================================
    
    #Ð·Ð°Ð´Ð°Ñ‡Ð° 1
    
    if parameter == 1:
        
        po[0] = 1.0
        u[0] = 1.0237498
        p[0] = 8.0
        po[n] = 0.1933880
        u[n] = 5.2937598
        p[n] = 0.8018469
    
    
    #Ð·Ð°Ð´Ð°Ñ‡Ð° 2 
    
    if parameter == 2:
        
        po[0] = 1.0
        u[0] = 1.0237498
        p[0] = 8.0
        po[n] = 0.8835893
        u[n] = 1.1586268
        p[n] = 7.0315580

    #=====================================

    for i in range(1,n): #???#
        po[i] = po[n]
        u[i] = u[n]
        p[i] = p[n]

    #     Ð¨ÐÐ“ ÐŸÐž Ð’Ð Ð•ÐœÐ•ÐÐ˜    #

    tau = 0.0001#0.4 * h/max(u)
    # 1)tau = 0.0001 + kappa = 0.5 (Ñ€Ð°Ð±Ð¾Ñ‚Ð°ÐµÑ‚)
    print(max(u),tau)  

    #---------------------- #    

    #Ð¿Ð¾ÑÐ»ÐµÐ¼ÐµÐ½Ñ‚Ð½Ð¾Ðµ ÑƒÐ¼Ð½Ð¾Ð¶ÐµÐ½Ð¸Ðµ Ð½Ð°Ð¼Ð¿Ð¸
    Q[0] = A*po
    Q[1] = A*po*u
    Q[2] = A*po*(u*u/2 + p/(po*(gamma-1))) 
    Qnext = Q.copy()

    F[0] = A*po*u
    F[1] = A*(po*u*u + p)
    F[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

    S[0] = 0
    S[1] = p * dA
    S[2] = 0

    Qs = Q.copy()
    Fs = F.copy()
    Ss = S.copy()
    
    cfl = 0.8
    time = 0
    #Ñ€Ð°ÑÑ‡ÐµÑ‚ Ð½Ð¾Ð²Ð¾Ð³Ð¾  q
    #Ñ€Ð°ÑÑ‡ÐµÑ‚ Ð½Ð¾Ð²Ñ‹Ñ… Ð´Ð°Ð²Ð»ÐµÐ½Ð¸Ñ ÑÐºÐ¾Ñ€Ð¾ÑÑ‚Ð¸ Ð¸ Ð¿Ð»Ð¾Ñ‚Ð½Ð¾ÑÑ‚Ð¸
    #Ñ€Ð°ÑÑ‡ÐµÑ‚ Ð½Ð¾Ð²Ñ‹Ñ… f and s
    for k in range(100000): #1) 1000

        V_max = 0
        for i in range(0, n):
            c = np.sqrt((np.abs(gamma*p[i]/po[i])))
            V_max = max(V_max, abs(u[i])+c)
            
        tau = cfl*h/V_max
        #Ð¿ÐµÑ€Ð²Ñ‹Ð¹ ÑˆÐ°Ð³
        Qnext[:,1:-1] = Q[:,1:-1] - tau/h * (F[:,2:] - F[:,1:-1]) + tau*S[:,1:-1]    

        poprev = po.copy()
        po[1:-1] = Qnext[0,1:-1]/A[1:-1]
        u[1:-1] = Qnext[1,1:-1]/(A[1:-1]*po[1:-1])
        p[1:-1] = (Qnext[2,1:-1]/(A[1:-1]*po[1:-1]) - u[1:-1]*u[1:-1]/2) * po[1:-1]*(gamma-1)

        M = u/c_0

        Qs[0] = Qnext[0].copy()
        Qs[1] = Qnext[1].copy()
        Qs[2] = Qnext[2].copy()

        Fs[0] = A*po*u
        Fs[1] = A*(po*u*u + p)
        Fs[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

        Ss[0] = 0
        Ss[1] = p * dA
        Ss[2] = 0

        #Ð²Ñ‚Ð¾Ñ€Ð¾Ð¹ ÑˆÐ°Ð³
        Qnext[:,1:-1] = 1/2*( Q[:,1:-1] + Qs[:,1:-1] - tau/h * (Fs[:,1:-1] - Fs[:,0:-2]) + tau*Ss[:,1:-1] )

        #poprev = po.copy()
        po[1:-1] = Qnext[0,1:-1]/A[1:-1]
        u[1:-1] = Qnext[1,1:-1]/(A[1:-1]*po[1:-1])
        p[1:-1] = (Qnext[2,1:-1]/(A[1:-1]*po[1:-1]) - u[1:-1]*u[1:-1]/2) * po[1:-1]*(gamma-1)

        M = u/c_0

        Q[0] = Qnext[0].copy()
        Q[1] = Qnext[1].copy()
        Q[2] = Qnext[2].copy()

        F[0] = A*po*u
        F[1] = A*(po*u*u + p)
        F[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

        S[0] = 0
        S[1] = p * dA
        S[2] = 0

        #Ð¿ÐµÑ€ÐµÑÑ‡ÐµÑ‚ Ñ ÑƒÑ‡ÐµÑ‚Ð¾Ð¼ Ð¸ÑÐºÑƒÑÑ‚Ð²ÐµÐ½Ð½Ð¾Ð¹ Ñ‡Ð¸ÑÐ»ÐµÐ½Ð½Ð¾Ð¹ Ð²ÑÐ·ÐºÐ¾ÑÑ‚Ð¸
        Qnext[:,1:-1] = Q[:,1:-1] + kappa*tau/h*( abs(u[2:]-u[1:-1])*(Q[:,2:] - Q[:,1:-1]) - abs(u[1:-1]-u[0:-2])*(Q[:,1:-1] - Q[:,0:-2]) )

        po[1:-1] = Qnext[0,1:-1]/A[1:-1]
        u[1:-1] = Qnext[1,1:-1]/(A[1:-1]*po[1:-1])
        p[1:-1] = (Qnext[2,1:-1]/(A[1:-1]*po[1:-1]) - u[1:-1]*u[1:-1]/2) * po[1:-1]*(gamma-1)

        M = u/c_0

        Q[0] = Qnext[0].copy()
        Q[1] = Qnext[1].copy()
        Q[2] = Qnext[2].copy()

        F[0] = A*po*u
        F[1] = A*(po*u*u + p)
        F[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

        S[0] = 0
        S[1] = p * dA
        S[2] = 0 

        eps = 0
        for i in range(0, n):
            eps = max(eps, abs((po[i] - poprev[i])/poprev[i])/tau)

        time += tau

        if eps<1e-4:
            print(tau)
            break
            
    return po, p, u, x
 
po1, p1, u1, x1 = Makk(500, 2)    
plt.title('Oscillations dt = ')   
plt.plot(x1, po1)
plt.xlabel('x')
plt.ylabel('po') 
```

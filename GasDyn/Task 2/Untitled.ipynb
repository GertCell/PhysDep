```python
#схема Лакса-Фридрихса
import math as mt
import numpy as np
import matplotlib.pyplot as plt


def Lax(n):
    #схема Лакса-Фридрихса в 2 шага
    
    eps = 0.0001
    gamma = 1.4
    L = 1.0
    
    a = 0
    b = L
    h = (b-a)/n

    #инициализация#
    x = np.linspace(0, L, n+1)
    A = np.linspace(0, L, n+1)
    dA = np.linspace(0, L, n+1)

    #xh = np.linspace(h, L-h, n) #на одну точку поменьше
    xh = np.linspace(-h, L+h, n+2) #больше?)
    Ah = np.linspace(-h, L+h, n+2) 
    dAh = np.linspace(-h, L+h, n+2) 

    po = np.linspace(0, L, n+1)
    poprev = np.linspace(0, L, n+1)
    p = np.linspace(0, L, n+1)
    M = np.linspace(0, L, n+1)
    u = np.linspace(0, L, n+1)

    poh = np.linspace(0, L, n)
    ph = np.linspace(0, L, n)
    uh = np.linspace(0, L, n)

    Q = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    F = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    S = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])
    Qnext = np.array([[1.1 for i in range(0,n+1)] for j in range(0,3)])

    Qh = np.array([[1.1 for i in range(0,n)] for j in range(0,3)]) #на одну точку поменьше
    Fh = np.array([[1.1 for i in range(0,n)] for j in range(0,3)])
    Sh = np.array([[1.1 for i in range(0,n)] for j in range(0,3)])

    As = 0.5
    xs = 0.5
    for i in range(0,n+1):
        A[i] = As + (1-As)*(1-x[i]/xs)**2 
    for i in range(1,n):
        dA[i] = 0.5 * (A[i+1] - A[i-1])/h  #4*x[i] - 2

    for i in range(0,n+2):
        Ah[i] = As + (1-As)*(1-xh[i]/xs)**2     
    for i in range(1,n+1):
        dAh[i] = 0.5 * (Ah[i+1] - Ah[i-1])/h  #4*x[i] - 2

    #нач и гран условия#

    #=====================================

    #задача 1

#     po[0] = 1.0
#     u[0] = 1.0237498
#     p[0] = 8.0
#     po[n] = 0.1933880
#     u[n] = 5.2937598
#     p[n] = 0.8018469

    #задача 2 

    po[0] = 1.0
    u[0] = 1.0237498
    p[0] = 8.0
    po[n] = 0.8835893
    u[n] = 1.1586268
    p[n] = 7.0315580

    #=====================================

    for i in range(1,n): #???#
        po[i] = po[0] + i * (po[n] - po[0])/n
        u[i] = u[0] + i * (u[n] - u[0])/n
        p[i] = p[0] + i * (p[n] - p[0])/n

    Sh[0] = 0
    Sh[2] = 0
    for i in range(0,n): 
        Sh[1,:] = p[i]        


    #       ШАГ ПО ВРЕМЕНИ     #

    tau = 0.1 * h/(round(max(u),0)) #2)
    #tau = 0.5 * h/(round(max(u),0)) #1)
    #tau = 0.00001
    print(tau, "tau")
    print(h, "h")
    print(n, "n")
    print("prob tau",0.8 * h/max(u))

    #-------------------------#

    #поэлементное умножение нампи
    Q[0] = A*po
    Q[1] = A*po*u
    Q[2] = A*po*( u*u/2 + p/(po*(gamma-1)) ) 
    Qnext = Q.copy()

    F[0] = A*po*u
    F[1] = A*(po*u*u + p)
    F[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

    S[0] = 0
    S[1] = p * dA
    S[2] = 0


    #расчет нового  q
    #расчет новых давления скорости и плотности
    #расчет новых f and s
    for k in range(10000000): #1) 1000
        poprev = po.copy()

        #первый шаг

        Qh[:,:] = 1/2 * (Q[:,0:-1] + Q[:,1:]) - tau/(2*h) * (F[:,1:] - F[:,0:-1]) + tau/2.0 * Sh[:,:]

        poh = Qh[0]/Ah[1:-1]
        uh = Qh[1]/(Ah[1:-1]*poh)
        ph = (Qh[2]/(Ah[1:-1]*poh) - uh*uh/2) * poh*(gamma-1)

        Fh[0] = Ah[1:-1]*poh*uh
        Fh[1] = Ah[1:-1]*(poh*uh*uh + ph)
        Fh[2] = Ah[1:-1]*uh*(poh*(uh*uh/2 + ph/(poh*(gamma-1))) + ph)

        Sh[0] = 0
        Sh[1] = ph * dAh[1:-1]
        Sh[2] = 0

        #второй шаг

        Qnext[:,1:-1] = 1/2 * (Qh[:,:-1] + Qh[:,1:]) - tau/(2*h) * (Fh[:,1:] - Fh[:,:-1]) + tau/2 * S[:,1:-1] #опечатка??

        po[1:-1] = Qnext[0,1:-1]/A[1:-1]
        u[1:-1] = Qnext[1,1:-1]/(A[1:-1]*po[1:-1])
        p[1:-1] = (Qnext[2,1:-1]/(A[1:-1]*po[1:-1]) - u[1:-1]*u[1:-1]/2) * po[1:-1]*(gamma-1)


        Q[0] = Qnext[0].copy()
        Q[1] = Qnext[1].copy()
        Q[2] = Qnext[2].copy()

        F[0] = A*po*u
        F[1] = A*(po*u*u + p)
        F[2] = A*u*(po*(u*u/2 + p/(po*(gamma-1))) + p)

        S[0] = 0
        S[1] = p * dA
        S[2] = 0


        M = u/np.sqrt(gamma * p/po)

        if ( max( abs((po-poprev)/(poprev* tau)) ) ) < eps:
            print(k,tau*k)
            break

    return po, p, u
      
Lax(100)
a, b, c = Lax(100)
plt.plot(a)
```

```python

import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
%matplotlib inline

Nx = 100
L = 1

T = np.zeros(Nx+1)

ai = 2
bi = 2
ci = 1
b_i = 2

a = -np.pi/2
b = np.pi/2

x = np.linspace(a, b, Nx + 1)
alfa = np.linspace(a,b,Nx + 1)
beta = np.linspace(a,b,Nx + 1)

hx = (b-a)/Nx

fi = lambda x_1: np.cos(x_1)*hx**2
fr = lambda x_2: -2/np.pi * x_2 + (1/2 + 2/np.pi)-np.cos(x_2)


alfa[0] = bi/(bi-ai*hx)
beta[0] = ci/(ai-bi/hx)

for i in range (1, Nx):

    alfa[i] = 1 / (b_i -  alfa[i - 1])
    beta[i] = (beta[i - 1] - fi(x[i])) / (b_i -  alfa[i - 1])

T[Nx] = (ci + bi/hx*beta[Nx-1]) / (ai + bi/hx - bi/hx*alfa[Nx-1])

for i in range(Nx-1,-1,-1):
    T[i] = alfa[i] * T[i + 1] + beta[i]
```

```python
plt.subplot(2,2,1)
plt.grid()
plt.plot(x, T)
plt.subplot(2,2,2)
plt.grid()
plt.plot(x, fr(x))
plt.subplot(2,2,3)
plt.grid()
plt.plot(x, fr(x) - T)
```

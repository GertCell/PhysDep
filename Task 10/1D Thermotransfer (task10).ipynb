```python

import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
%matplotlib inline



#entering constants

Nx = 50
Ny = 50

t_end = 5
Ly = 0.05
Lx = 1
l = 18
po = 4500
c = 530
T0 = 200
Th = 1500
Tc = 200

# summation steps

a = l/(po*c)
hx = Lx/(Nx-1)
hy = Ly/(Ny-1)
tau = t_end/100

#Temp arrays
T = np.zeros(Nx+1)
alfa = np.zeros(Nx)
beta = np.zeros(Ny)

for i in range(0, Nx+1):
    T[i] = T0
T[0] = Th

time = 0
t = -1

fig, ax = plt.subplots(1, 1, figsize=(12, 4), constrained_layout=True)
```

```python



Nx = 50
Ny = 50

t_end = 0.5

Lx = 1

po = 1
c = 1
T0 = 0
Th = 0
Tc = 0
l = 1

T = np.zeros(Nx+1)
alfa = np.zeros(Nx)
beta = np.zeros(Ny)
L = 1
x = np.linspace(0, L, Nx+1)
for i in range(0, Nx+1):
    T[i] = x[i]*(1-x[i]/L)**2
T[0] = Th


time = 0
t_end = 0.5

tau = t_end/100
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
plt.grid()
plt.plot(T)

while time <= t_end:

    T[0] = 0

    time = time + tau

    alfa[0] = 0
    beta[0] = 0

    for i in range (1, Nx):

        ai = l / hx**2
        bi = 2 * l / hx**2 + po * c / tau
        fi = -po * c * T[i] / tau + tau /2*(T[i+1]-2*T[i]+T[i-1])/hx**2

        alfa[i] = ai / (bi - ai * alfa[i - 1])
        beta[i] = (ai * beta[i - 1] - fi) / (bi - ai * alfa[i - 1])

    T[Nx] = 0
    for i in range (0, Nx):
        T[i] = alfa[i] * T[i + 1] + beta[i]
    if time <= 0.04:
        plt.plot(T)
    else: break
```

```python
from matplotlib.pyplot import *
import numpy as np
%matplotlib notebook

a = 998
b = 1998
c = -999
d = -1999

fx = lambda t, x, y: a * x + b * y

fy = lambda t, x, y: c * x + d * y

t0 = 0
x0 = 1
y0 = 1

X = [x0]
Y = [y0]
T = [t0]

T_h = 1
N = 600  
h = T_h / N
print(h)



subplot(111)
axis([0, 1, 0, 5])
ln1, = plot(T, X)




def MethodKramer(f1, f2):
    delta = (h*a-1)*(h*d - 1) - h*b*h*c
    delta1 = -f1*(h*d - 1) + f2*h*b
    delta2 = (h*a - 1)*(-f2) + h*c*f1
    
    return delta1/delta, delta2/delta



def rk2():
    
    t = t0
    for i in range(0,N):
        x, y = MethodKramer(X[i], Y[i])
        
        t = t + h
        T.append(t)
        X.append(x)
        Y.append(y)
        
        ln1.set_data(T, X)
        gcf().canvas.draw()
    return 0

rk2()

```

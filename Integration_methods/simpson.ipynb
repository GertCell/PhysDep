```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *

%matplotlib notebook


def f(x):
    return x * (1 - x)

def f1(x):
    return 1/x/x

a = 0
b = 1
v = 1 / 6  # np.pi/2
S1 = []
I = []
Y = []
N = []

def simpson():
    
    subplot(111)
    ln, = plot(I, S1)
    ln1, = plot(N, Y)
    axis([0, 50, 0, 1])
    
    for n in range(2, 50, 2):
        
        d = (b - a) / n
        xi = a
        s1 = 0
        
        for i in range(1, n, 2):
            
            s1 = s1 + 4 * f(xi + i * d) + 2 * f(xi + (i + 1) * d)
        
        s1 = s1 - f(b) * 2
        S = d / 3 * (s1 + f(a) + f(b))
        S1.append(v - S)
        I.append(n)
        
        ln.set_data(I, S1)
        gcf().canvas.draw()
        
    for i in arange(0.1, 50, 0.5):
        Y.append(f1(i))
        N.append(i)
        
        ln1.set_data(N, Y)
        gcf().canvas.draw()
        
    return 0

simpson()
```

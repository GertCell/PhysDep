```python
from numpy import *
from matplotlib.pyplot import *
from scipy.integrate import *

%matplotlib notebook


def f(x):
    return x**(1/3)*(exp(sin(x)))

def f1(x):
    return 1/((x))

a = 0
b = 1
v, err = quad(f, a, b)#np.pi/2
print(v)
S1 = []
I = []
Y = []
N = []

def simpson():
    
    subplot(111)
    ln, = plot(I, S1)
    ln1, = plot(N, Y)
    axis([0, 50, 0, 1e-2])
    
    for n in range(2, 50, 2):
        
        d = (b - a) / n
        xi = a
        s1 = 0
        
        for i in range(1, n, 2):
            
            s1 = s1 + 4 * f(xi + i * d) + 2 * f(xi + (i + 1) * d)
        
        s1 = s1 - f(b)*2
        S = d / 3 * (s1 + f(a) + f(b))
        print(S)
        S1.append(abs(v - S))
        I.append(n)
        
        ln.set_data(I, S1)
        gcf().canvas.draw()
        
    for i in arange(0.1, 50, 0.5):
        Y.append(f1(i))
        N.append(i)
        
        ln1.set_data(N, Y)
        gcf().canvas.draw()
        
    return 0

def trapz():
    
    subplot(111)
    ln3, = plot(I, S1)
    ln1, = plot(N, Y)
    axis([0, 50, 0, 1])
    
    for n in range(2, 50, 2):
        d = (b - a) / n
        xi = a
        xi_1 = xi + d
        s1 = (f(xi_1) + f(xi)) * d
        for i in range(0, n - 1):
            xi = xi_1
            xi_1 = xi + d
            s = s1
            s1 = s + (f(xi_1) + f(xi)) * d
            
        
        S1.append(abs(v - s/2))
        I.append(n)
        ln3.set_data(I, S1)
        gcf().canvas.draw()
    
    return 0

simpson()
trapz()
```
